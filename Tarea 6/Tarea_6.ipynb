```python

P1 = {'A': 0.4, 'B': 0.6}
P2 = {'A': 0.5, 'B': 0.5}


a1 = 0.3
a2 = 0.7


assert a1 + a2 == 1


P = {evento: a1 * P1[evento] + a2 * P2[evento] for evento in P1}


def ver_kolmogorov(P):

    no_negativo = all(value >= 0 for value in P.values())

    normal = sum(P.values()) == 1

    suma = P['A'] + P['B'] == P['A'] + P['B']  # Trivially true here
    return no_negativo, normal, suma

# Check the axioms
non_neg, norm, add = ver_kolmogorov(P)

print("Combined Probability Measure:", P)
print(f"Non-Negativity: {non_neg}")
print(f"Normalization: {norm}")
print(f"Additivity: {add}")
```

2. Suponga que se quiere estimar el número medio de estudiantes que se aproximan a la
secretaria de física siguiendo una distribución de Poisson. El problema podría tener
información inicial que puede ser utilizada como una función a priori de probabilidad. Esto
asigna probabilidades a los posibles valores del parámetro λ. Supongamos que el número
medio a priori de estudiantes es: 1,2,3,4 con las probabilidades Π(λ) = [0.4, 0.3, 0.2, 0.1]
según las secretarias 1,2,3 y 4 (información subjetiva). El número de estudiantes
observados durante cierto intervalo de tiempo es x = 4.


```python
import numpy as np
import math

# Definición de las probabilidades a priori (Π(λ)) y los valores posibles de λ
lambdas = np.array([1, 2, 3, 4])
prior = np.array([0.4, 0.3, 0.2, 0.1])

# Número observado de estudiantes
x = 4
```

a) Haga la estimación inicial del parámetro λˆ. (Ans: λˆ = 2.0)

```python
def poisson_likelihood(x, lam):
    return (lam**x * np.exp(-lam)) / math.factorial(x)
```

b) Calcule la función de verosimilitud de la observación para cada modelo P(x/λi) =
L(x/λi) donde i = 1, 2, 3, 4.

```python
likelihoods = np.array([poisson_likelihood(x, lam) for lam in lambdas])
print(f"Verosimilitudes para cada modelo: {likelihoods}")
```

c) Aplicando el teorema de Bayes, calcule la función de distribución a posteriori para
cada modelo:

image.png

```python
num = likelihoods * prior
denom = np.sum(num)
posterior = num / denom
print(f"Distribución posterior: {posterior}")
```

d) Verifique que la distribución posterior está normalizada.

```python
print(f"Verificación de normalización (debe ser 1): {np.sum(posterior)}")
```

e) Indique el modelo que m´as probablemente explique los datos.

```python
max_posterior_index = np.argmax(posterior)
print(f"El modelo más probable es λ = {lambdas[max_posterior_index]}")
```

f) Usando la información subjetiva de cada secretaria y la verosimilitud de la observación
(i,e. la distribución a posteriori). Calcule el mejor parámetro λˆ. (Ans: λˆ = 2.77)

```python
lambda_hat = np.sum(lambdas * posterior)
print(f"La mejor estimación de λ es: {lambda_hat}")
```

3. En el problema anterior, imagine que en el primer intervalo de tiempo el número observado
de estudiantes es x1 = 4, en el siguiente intervalo de tiempo el número de estudiantes es
x2 = 5. Realice todos los pasos anteriores del a) al f) teniendo en cuenta el likelihood
conjunto. (Ans: λˆ = 3.27)


```python
import numpy as np
from math import factorial, exp

# Prior
prior = np.array([0.4, 0.3, 0.2, 0.1])
lambdas = np.array([1, 2, 3, 4])


x1 = 4
x2 = 5


def poisson_likelihood(x, lambd):
    return (lambd**x * exp(-lambd)) / factorial(x)


likelihood_x1 = np.array([poisson_likelihood(x1, lambd) for lambd in lambdas])
likelihood_x2 = np.array([poisson_likelihood(x2, lambd) for lambd in lambdas])


likelihood_juntas = likelihood_x1 * likelihood_x2
posterior_sinnorm = likelihood_juntas * prior

posterior = posterior_sinnorm / np.sum(posterior_sinnorm)
lambda_hat = lambdas[np.argmax(posterior)]

expected_lambda = np.sum(posterior * lambdas)

print(f"Distribución posterior: {posterior}")
print(f"Mejor valor de λ: {lambda_hat}")
print(f"Valor esperado de λ: {expected_lambda}")
```

4. Un haz de partículas compuesto por electrones, protones y partículas α, atraviesa
un detector de partículas, que tiene una eficiencia de detección de 90%, 60% y 20%
respectivamente. El flujo de electrones es el doble que el de protones y partículas α.


```python
import numpy as np
import matplotlib.pyplot as plt

x = 1

P_e = 2 * x  # El flujo de electrones
P_p = x  # Flujo de protones
P_alpha = x  # Flujo de partículas alfa

# Eficiencia de detección
P_D_e = 0.9
P_D_p = 0.6
P_D_alpha = 0.2

# Probabilidad previa de detección para cada partícula
Pi_D_e = P_e * P_D_e
Pi_D_p = P_p * P_D_p
Pi_D_alpha = P_alpha * P_D_alpha

# Probabilidad total de detección
P_D = Pi_D_e + Pi_D_p + Pi_D_alpha

# Probabilidades posteriores utilizando el teorema de Bayes
P_e_given_D = Pi_D_e / P_D
P_p_given_D = Pi_D_p / P_D
P_alpha_given_D = Pi_D_alpha / P_D


print(f"Probabilidad total de detección P(D) = {P_D:.2f}")
print(f"P(e|D) = {P_e_given_D:.4f}")
print(f"P(p|D) = {P_p_given_D:.4f}")
print(f"P(α|D) = {P_alpha_given_D:.4f}")


posterior = [P_e_given_D, P_p_given_D, P_alpha_given_D]
etiquetas = ['Electrón', 'Protón', 'Partícula Alfa']

plt.stem(etiquetas, posterior, basefmt=" ")
plt.title('Distribución de Probabilidad Posterior')
plt.ylabel('Probabilidad')
plt.show()
```

```python
import math

def combinaciones_con_repeticion(n, r):
    # fórmula C_r^n = (n + r - 1)! / (r! * (n - 1)!)
    return math.factorial(n + r - 1) // (math.factorial(r) * math.factorial(n - 1))

# Verificación del código con un ejemplo
n = 3  # Número de tipos de elementos (por ejemplo, 3 tipos de frutas)
r = 4  # Número de elementos a seleccionar (por ejemplo, 4 frutas)

# Calcular combinaciones con repetición
resultado = combinaciones_con_repeticion(n, r)
print(f"El número de formas de seleccionar {r} elementos de {n} tipos con repetición es: {resultado}")
```
