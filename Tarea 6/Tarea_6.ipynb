3. En el problema anterior, imagine que en el primer intervalo de tiempo el nmero observado
de estudiantes es x1 = 4, en el siguiente intervalo de tiempo el nmero de estudiantes es
x2 = 5. Realice todos los pasos anteriores del a) al f) teniendo en cuenta el likelihood
conjunto. (Ans:  = 3.27)


```python
import numpy as np
from math import factorial, exp

# Prior
prior = np.array([0.4, 0.3, 0.2, 0.1])
lambdas = np.array([1, 2, 3, 4])


x1 = 4
x2 = 5


def poisson_likelihood(x, lambd):
    return (lambd**x * exp(-lambd)) / factorial(x)


likelihood_x1 = np.array([poisson_likelihood(x1, lambd) for lambd in lambdas])
likelihood_x2 = np.array([poisson_likelihood(x2, lambd) for lambd in lambdas])


likelihood_juntas = likelihood_x1 * likelihood_x2
posterior_sinnorm = likelihood_juntas * prior

posterior = posterior_sinnorm / np.sum(posterior_sinnorm)
lambda_hat = lambdas[np.argmax(posterior)]

expected_lambda = np.sum(posterior * lambdas)

print(f"Distribucin posterior: {posterior}")
print(f"Mejor valor de : {lambda_hat}")
print(f"Valor esperado de : {expected_lambda}")
```

4. Un haz de partculas compuesto por electrones, protones y partculas , atraviesa
un detector de partculas, que tiene una eficiencia de deteccin de 90%, 60% y 20%
respectivamente. El flujo de electrones es el doble que el de protones y partculas .


```python
import numpy as np
import matplotlib.pyplot as plt

x = 1

P_e = 2 * x  # El flujo de electrones
P_p = x  # Flujo de protones
P_alpha = x  # Flujo de partculas alfa

# Eficiencia de deteccin
P_D_e = 0.9
P_D_p = 0.6
P_D_alpha = 0.2

# Probabilidad previa de deteccin para cada partcula
Pi_D_e = P_e * P_D_e
Pi_D_p = P_p * P_D_p
Pi_D_alpha = P_alpha * P_D_alpha

# Probabilidad total de deteccin
P_D = Pi_D_e + Pi_D_p + Pi_D_alpha

# Probabilidades posteriores utilizando el teorema de Bayes
P_e_given_D = Pi_D_e / P_D
P_p_given_D = Pi_D_p / P_D
P_alpha_given_D = Pi_D_alpha / P_D


print(f"Probabilidad total de deteccin P(D) = {P_D:.2f}")
print(f"P(e|D) = {P_e_given_D:.4f}")
print(f"P(p|D) = {P_p_given_D:.4f}")
print(f"P(|D) = {P_alpha_given_D:.4f}")


posterior = [P_e_given_D, P_p_given_D, P_alpha_given_D]
etiquetas = ['Electrn', 'Protn', 'Partcula Alfa']

plt.stem(etiquetas, posterior, basefmt=" ")
plt.title('Distribucin de Probabilidad Posterior')
plt.ylabel('Probabilidad')
plt.show()
```
