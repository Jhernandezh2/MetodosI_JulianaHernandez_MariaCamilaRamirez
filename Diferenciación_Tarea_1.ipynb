# **EJERCICIOS DE DERIVACIÓN: 2**

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.05

def f(x):
    return 1 / np.sqrt(1 + np.exp(-x**2))

def central_difference(x:int, h:float)->float:
    return (f(x + h) - f(x - h)) / (2 * h)

def g(x):
  #return (x*np.exp(-x**2))/(((np.exp(-x**2)+1))**(3/2))
  return (x * np.exp(-x**2)) / ((1 + np.exp(-x**2)) * np.sqrt(1 + np.exp(-x**2)))

x_values = np.linspace(-10,10,400)
h = 0.05

central_derivative_values = central_difference(x_values, h)
regular_derivative_values = g(x_values)
errors = np.abs(central_derivative_values - regular_derivative_values)

#plt.plot(x_values, central_derivative_values)
#plt.plot(x_values, regular_derivative_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, central_derivative_values, label="Central Difference Derivative", color="blue")
plt.plot(x_values, regular_derivative_values, label="Analytical Derivative", color="k",ls=(5,(5,5)))
plt.plot(x_values, errors, label="Error", color="red", linestyle="--")
```

# **MÁXIMOS**

```python
import numpy as np
import requests
import matplotlib.pyplot as plt


url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
response = requests.get(url)
data = response.text.strip()

x_values = []
y_values = []
for line in data.splitlines():
    x, y = map(float, line.split())
    x_values.append(x)
    y_values.append(y)

x_values = np.array(x_values)
y_values = np.array(y_values)

def find_local_maxima(x, y):
    max_x = []
    max_y = []
    for i in range(1, len(y) - 1):
        if y[i] > y[i - 1] and y[i] > y[i + 1]:
            max_x.append(x[i])
            max_y.append(y[i])
    return max_x, max_y


max_x, max_y = find_local_maxima(x_values, y_values)


plt.scatter(max_x,max_y,color='red',label="Maximos locales")

plt.plot(x_values,y_values,label="Datos")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.legend()
plt.grid(True)
plt.show()
```

# **SECUENCIA DE FIBONACCI**

```python
import matplotlib.pyplot as plt
import numpy as np


#secuencia_de_fibonacci = [0,1]
n = 20
x_values_n = np.arange(0,n,1)

numero_aureo = (1+np.sqrt(5))/2

def generar_secuencia_fibonacci(n):
  secuencia_de_fibonacci = [0, 1]
  for i in range (2,n):
    siguiente = secuencia_de_fibonacci[i-1] + secuencia_de_fibonacci[i-2]
    secuencia_de_fibonacci.append(siguiente)

  return secuencia_de_fibonacci

plt.plot(x_values_n,generar_secuencia_fibonacci(n))
plt.show()

x_values_phi = np.arange(0,18,1)

secuencia_de_fibonacci = generar_secuencia_fibonacci(n)
fib_valores = [secuencia_de_fibonacci[i+1] / secuencia_de_fibonacci[i] for i in range(1, len(secuencia_de_fibonacci)-1)]


plt.plot(x_values_phi, fib_valores, color='blue', label="Secuencia de Fibonacci")
plt.axhline(y=numero_aureo, color='red', linestyle='--', label="Número áureo")

plt.xlabel("n")
plt.ylabel("f(n+1)/f(n)")
plt.title("Comparación de secuencia de Fibonaci al número áureo")
plt.legend()
plt.grid(True)


plt.show()

```

# **ESPIRAL DE ARQUÍMEDES**

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 1

angulo = np.linspace(0, 2 * np.pi, 500)
r = a + b * angulo

x = r * np.cos(angulo)
y = r * np.sin(angulo)

plt.plot(x, y)
plt.title('Espiral de Arquímedes')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
