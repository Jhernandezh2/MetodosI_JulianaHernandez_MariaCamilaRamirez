 **Punto 5: Raíces de los primeros 5 polinomios de Laguerre**

image.png

```python
import sympy as sp

x = sp.Symbol('x')


def laguerre_poli(i):

    f = sp.exp(-x) * x**i

    derivada = sp.diff(f, x, i)

    L_i = derivada / sp.factorial(i)

    return L_i

def raices_laguerre(i):

    L_poli = laguerre_poli(i)
    raices = sp.solvers.solve(L_poli, x)

    raices_reales = [r.evalf() for r in raices if r.is_real]
    return raices_reales

ordenes = [1, 2, 3, 4, 5]


for i in ordenes:
    raices = raices_laguerre(i)
    print(f"Raíces reales del polinomio de Laguerre L_{i}(x): {raices}")
```
